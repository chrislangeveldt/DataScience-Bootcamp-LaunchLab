## Plotting distributions
Next we will analyze the distribution of each feature, through a histogram

Finish the code below to plot histograms for each of the following features: *MPG, Cylinders, Displacement, Horsepower, Weight, Acceleration, Model Year, Origin*

For continuous features, use the default binsize of matplotlib.

```python
plt.style.use('seaborn-whitegrid')
fig, axs = plt.subplots(4,2,figsize=(10,20))
fig.suptitle('Histogram of each feature.')

# insert code here
columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin']
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet', 'purple']
bins = [10, 5, 10, 10, 10, 10, 10, 3]
for i in range(4):
    for j in range(2):
        axs[i, j].hist(df[columns[i*2+j]], color=colors[i*2+j], bins=bins[i*2+j])
        axs[i, j].set_xlabel(columns[i*2+j])
        axs[i, j].set_ylabel('# of observations')


fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
```

# Emphasising correlation

## Complete the code below to plot a scatterplot for MPG vs Weight.

Furthermore, indicate using colours the different origins of each point.

```python
fig, ax = plt.subplots(figsize=(7,6))

colors = {'USA':'tab:blue', 'JAPAN':'tab:red', 'EUROPE':'tab:green'}

# insert code here
for country in colors:
    ax.scatter(df[df['origin']==country]['mpg'], df[df['origin']==country]['weight'], color=colors[country], label=country)
ax.legend()
ax.set_xlabel('mpg')
ax.set_ylabel('weight')

plt.show()
```

### Does MPG and Weight have positive, negative or no correlation
*negative correlation*

# Working with time series
Next we will be working with a new dataset: 

The daily temperature for an arbitrary location was recorded from 1981 to 1991.

Once again answer the following questions plotting only with matplotlib.

Hint: For some of the following questions it is necessary to restructure the dataframe first using pd.groupby(). Seperating each year into a new dataframe column.


```python
# Load in data
df = pd.read_csv('src/temps.csv', index_col=0, parse_dates=True, header=0, squeeze=True)
df.head()
```

## Plot the daily temperature for each recorded measurement.

```python
fig, ax = plt.subplots(figsize=(12,5))

# insert code here
ax.plot(df.index, df, color='r', linewidth=0.3, marker='o', markersize=2, markerfacecolor='r', markeredgecolor='r')
ax.set_xlabel('Daily')
ax.set_ylabel('Temperature')
ax.set_title('Temperature daily')

plt.show()
```

## Comparing different years
Complete the code below to plot the temperature of each year on a different subplot.

```python
fig, ax = plt.subplots(10,1,figsize=(10,20))

# insert code here
for i in range(10):
    year = 1981 + i
    begin = str(year) + '-01-01'
    end = str(year) + '-12-31'
    data = df[begin:end]
    ax[i].plot(data.index, data, color='r', linewidth=0.3, marker='o', markersize=2, markerfacecolor='r', markeredgecolor='r')
    ax[i].set_xlabel('Daily')
    ax[i].set_ylabel('Temperature')
    ax[i].set_title('Temperature daily for year ' + str(year))


fig.suptitle('Daily Temperature Recorded for each respective year')
plt.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
```

## Comparing distributions of time intervals
To properly visualise and compare the distributions of yearly or monthly data, we employ the use of boxplots.

### Write code below to plot a boxplot for the temperatures of each respective year.

```python
fig, ax = plt.subplots(figsize=(10,5))

# insert code here
data = []
labels = []
for i in range(10):
    year = 1981 + i
    begin = str(year) + '-01-01'
    end = str(year) + '-12-31'
    data.append(df[begin:end].tolist())
    labels.append(str(year))
    
ax.boxplot(data)
ax.set_xticklabels(labels)
ax.set_xlabel('Year')
ax.set_ylabel('Temperature')
ax.set_title('Boxplot of temperatures per year')
    
plt.tight_layout()
plt.show()
```

### If answered correctly, we can see that 1982 had the largest spread in recorded temperature. 
We therefore seek to further analyse this specific year. Finish the code below to plot boxplots for each month in 1982.


```python
fig, ax = plt.subplots(figsize=(10,5))
x_labels = ['Jan', 'Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
ndays = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# insert code here
data = []
for i in range(12):
    month = 1 + i
    zero = ''
    if month < 10:
        zero = '0'
    begin = '1982-' + zero + str(month) + '-01'
    end = '1982-' + zero + str(month) + '-' + str(ndays[i])
    data.append(df[begin:end].tolist())
        
ax.boxplot(data)
ax.set_xticklabels(x_labels)
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
ax.set_title('Boxplot of temperatures per month of 1982')
    
plt.tight_layout()
plt.show()
```
